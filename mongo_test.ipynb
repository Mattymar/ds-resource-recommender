```python
data = [post['text'] for post in test_posts.find()[:300]]
```

```python
count_vectorizer = CountVectorizer(ngram_range=(1, 2),
                                   stop_words='english',
                                   token_pattern="\\b[a-z][a-z]+\\b")

X = count_vectorizer.fit_transform(data)
```

```python
X.shape
```

```python
type(X)
```

```python
n_topics = 10
n_iter = 10
lda = LatentDirichletAllocation(n_topics=n_topics, max_iter=n_iter)
lda.fit(X)
```

```python
def print_topic_top_words(model, cv, n_top_words=10):
    feature_names = cv.get_feature_names()
    
    for topic_vec in enumerate(model.components_):
        topic_num = topic_vec[0]
        topic_words = topic_vec[1]
        
        print('Topic {}:'.format(topic_num + 1))
        
        topic_values = sorted(zip(topic_words, feature_names), reverse=True)[:n_top_words]
        
        print(', '.join([y for x,y in topic_values]))
    return
```

```python
n_top_words = 20

print_topic_top_words(lda, count_vectorizer, n_top_words)
```
